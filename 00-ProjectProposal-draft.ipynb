## Research Question



To what degree can NBA player workload features like game frequency, minutes played, and travel distance be used to predict injuries in the NBA, and which variables contribute the most to higher risk?


## Background and Prior Work

NBA injuries are important to study because they affect player health and player availability. When players get hurt, they may miss games or require changes to team decisions about player usage. Mack et al. analyzed NBA injuries from the 2013-2014 through 2018-2019 seasons and found between 1,550 and 1,892 injuries per season, with 33.6% to 38.5% of those injuries resulting in at least one missed NBA game. Their study also reported game-loss injury rates per 10,000 player-minutes, which makes minutes played a relevant workload variable for studying injury risk.<a name="cite_ref-bg-1"></a>[<sup>1</sup>](#cite_note-bg-1)

Our research question focuses on workload because NBA players go through a long regular season with frequent games, high minutes, travel, and limited recovery time. Workload can include how many minutes a player plays, how often they play, how many rest days they get, and how far they travel between games. The NBA has also recognized that schedule congestion can affect player recovery. For the 2017-2018 season, the league extended the regular-season schedule window, eliminated four-games-in-five-nights stretches, and reduced back-to-back games from 16.3 to 14.4 per team.<a name="cite_ref-bg-2"></a>[<sup>2</sup>](#cite_note-bg-2) This supports studying rest days, game frequency, minutes played, and travel distance as possible predictors of injury.

Previous NBA research shows that the connection between schedule difficulty and injuries is not simple. Teramoto et al. studied NBA game injuries from the 2012-2013 through 2014-2015 regular seasons and examined schedule factors such as back-to-back games, four games in five days, and away games. They found that playing back-to-back games or four games in five days alone was not associated with a higher rate of game injuries, but playing away from home was positively associated with game injuries. They also found that back-to-back games and away games were significant predictors of players having frequent game injuries.<a name="cite_ref-bg-3"></a>[<sup>3</sup>](#cite_note-bg-3) This matters for our project because it suggests we should compare several workload features together instead of assuming that one factor, such as back-to-backs, explains injury risk by itself.

Other basketball research suggests that recent workload can be studied relative to what a player is normally used to. Weiss et al. studied professional basketball players using the acute:chronic workload ratio, which compares a player’s current weekly workload to their average workload from the previous four weeks. They reported that fewer team members were injured after workload ratios between 1.0 and 1.49 than after very low, low, or high workload ratios, although they also noted that the mean differences in injury likelihood were unclear and that workload-injury trends varied by player.<a name="cite_ref-bg-4"></a>[<sup>4</sup>](#cite_note-bg-4) This supports using recent workload features, such as minutes played in the last five games or games played in the last 7, 14, or 30 days, while still being careful not to treat workload as a simple one-variable cause of injury.

Travel is also relevant because it can affect sleep, fatigue, and recovery. Charest et al. studied NBA back-to-back games from 2013-2020 and looked at travel distance, travel direction, and game order. Their study focused on winning percentage rather than injuries, but they found that travel distance and travel patterns were related to performance outcomes. They also explained that travel fatigue and circadian disruption can disturb sleep and recovery across an NBA season.<a name="cite_ref-bg-5"></a>[<sup>5</sup>](#cite_note-bg-5) For our project, this supports including travel distance as a possible injury-risk feature, while being careful not to claim that travel distance directly causes injuries unless our own analysis supports that conclusion.

An especially similar prior project is Cohan et al.’s NBA injury prediction model. They used public injury data, player-game data, game activity, and player statistics to predict future injuries. Their dataset included about 238,000 player-game observations, but only about 5,000 injuries were associated with games, meaning injuries were rare compared with non-injury observations. They also found that high recent playing-time load, defined as more than 100 minutes in the previous five games, influenced injury incidence.<a name="cite_ref-bg-6"></a>[<sup>6</sup>](#cite_note-bg-6) Our project builds on this idea by focusing on easy-to-understand workload features, such as recent minutes, rest days, game frequency, and travel distance, and testing which of those variables contribute most to higher injury risk.
